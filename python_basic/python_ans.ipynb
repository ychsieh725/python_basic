### 有一筆學生資料如下:

```python
students_grade = [
    {"name": "Denji", "國文": 30, "英文": 10, "數學": 20},
    {"name": "Power", "國文": 60, "英文": 85, "數學": 70},
    {"name": "Aki", "國文": 95, "英文": 88, "數學": 92},
    {"name": "Pochita", "國文": 50, "英文": 65, "數學": 20},
    {"name": "Makima", "國文": 100, "英文": 98, "數學": 95},
    {"name": "Reze", "國文": 100, "英文": 40, "數學": 30},
]
```

##### 1. 印出所有學生姓名

```python
for student_grade in students_grade:
    # print(student_grade, end=", ")
    print(student_grade["name"], end=", ")
```

##### 2. 印出 `Pochita` 的數學成績

```python
for student_grade in students_grade:
    if student_grade["name"] == "Pochita":
        print(student_grade["數學"])
```

#### 3. 計算每位學生的總分
輸出格式  
`Phochita 總分: xxx 分`

```python
for student_grade in students_grade:
    # print(student_grade)
    total = 0
    for key, value in student_grade.items():
        if key == "name": # 當 key = name 的時候要跳過 不需要加入總分
            continue
        total += value
    
    print(f"{student_grade["name"]} 總分 {total} 分")
    
    # break
```

```python
def get_grade_total(student_grade: dict)-> int: # hints
    total = 0
    for key, value in student_grade.items():
        if key == "name": # 當 key = name 的時候要跳過 不需要加入總分
            continue
        total += value

    return total
```

```python
for student_grade in students_grade:

    total = get_grade_total(student_grade)
    
    print(f"{student_grade["name"]} 總分 {total} 分")
```

#### 4. 計算每位學生的平均分數
輸出格式  
`Pochita 平均分數: xxx 分`

```python
for student_grade in students_grade:

    total = get_grade_total(student_grade)
    
    print(f"{student_grade["name"]} 平均分數 {total/3:.2f} 分")
```

#### 5. 找出數學最高分的學生

```python
print(students_grade)
students_grade.sort(key=lambda x: x["數學"], reverse=True)
print(students_grade[0]["name"])
```

#### 6. 計算全班國文平均

```python
chi_total = 0
for student_grade in students_grade:
    chi_total += student_grade["國文"]

print(f"國文平均: {chi_total/len(students_grade):.2f} ")
```

#### 7. 找出不及格的學生（任一科 < 60）

```python
for student_grade in students_grade:
    for key, value in student_grade.items(): # 取得每個人的成績
        if key == "name":
            continue
        
        if value < 60:
            print(student_grade["name"])
            break # 一科不及格就打印 只需要打印一次
```

#### 8. 新增「總分」到每個 dict
例如 :  
`{"name": "Pochita", "國文": 80, ..., "總分": 245}`

```python
def get_grade_total(student_grade: dict)-> int: # hints
    total = 0
    for key, value in student_grade.items():
        if key == "name": # 當 key = name 的時候要跳過 不需要加入總分
            continue
        total += value

    return total

result = []
for student_grade in students_grade:
    total = get_grade_total(student_grade)
    student_grade["總分"] = total
    result.append(student_grade)

print(result)
```

#### 9. 找出三科都 > 80 的學生

```python
for student_grade in students_grade:
    all_pass = True
    for key, value in student_grade.items(): # 取得每個人的成績
        if key == "name":
            continue
        
        if value < 80:
            all_pass = False
            break # 一科低於80就跳過

    if all_pass:
        print(student_grade["name"])
    
```

#### 10. 將資料轉成以下格式 (dict of dict)

```python
"""
{
    "Denji": {"國文": 30, "英文": 10, "數學": 20},
    .....
}
"""
result = {}

for student_grade in students_grade:
    result[student_grade["name"]] = {}
    print(result)
    for key, value in student_grade.items():
        if key == "name":
            continue

        result[student_grade["name"]][key] = value

print(result)
```
