```python
# 匯入套件
import json, os

# 檔案路徑
folder_path = "../Module7" # 指定參數
files = os.listdir(folder_path)
print(files) #列出目前檔案路徑

# 如果指定路徑的檔案存在，則執行刪除
file_path = "data.txt"
print(os.path.exists(file_path))
if os.path.exists(file_path):
    os.remove(file_path) # 謹慎使用
```

```python
# os.mkdir()建立資料夾
folder = "test"
if not os.path.exists(folder):
    os.mkdir(folder)

# 資料夾裡面寫入新檔案
with open (f"{folder}/123.json", "w") as f:
    json.dump([1, 2, 3], f)
```
