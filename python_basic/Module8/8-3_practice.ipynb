##### 將 data.txt 的資料分類成另外三個檔案，分別為 cc.txt 、 css.txt 、 c.txt  

`我的解法`

```python
with open("data.txt", "r") as data:
    # 使用 strip() 去除每行末尾的換行符號 \n
    text_list = [line.strip() for line in data.readlines()]

cc = []
css = []
c = []

for i in text_list:
    if i[7:] == ".cc":
        cc.append(i)
    elif i[7:] == ".css":
        css.append(i)
    elif i[7:] == ".c":
        c.append(i)

cc_result = "\n".join(cc)
css_result = "\n".join(css)
c_result = "\n".join(c)

with open (file = "cc.txt", mode = "w", encoding="utf-8") as cc_write:
    cc_write.write(cc_result)

with open (file = "css.txt", mode = "w", encoding="utf-8") as css_write:
    css_write.write(css_result)

with open (file = "c.txt", mode = "w", encoding="utf-8") as c_write:
    c_write.write(c_result)
```

`老師解法`

```python
with open("data.txt", "r") as text:
    text_list = text.readlines()


for i in text_list:
    i = i.strip("\n") # 判斷前先把 \n 拿掉
    if i[7:] == ".cc":
        with open("cc.txt", "a")as f:
            f.write(i+"\n")
    elif i[7:] == ".css":
        with open("css.txt", "a")as f:
            f.write(i+"\n")
    elif i[7:] == ".c":
        with open("c.txt", "a")as f:
            f.write(i+"\n")
```
