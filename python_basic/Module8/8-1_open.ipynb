##### 讀取檔案 read( )

```python
# 印出 50 個 "="
print("="*50)

# 讀取檔案內容 - 常用的寫法 (執行完畢會自動關閉檔案)
with open(file = "../Module2/2-1.py", mode = "r", encoding="utf-8") as f:
    texts = f.read()
    # 一離開縮排，就關閉檔案
    # 2-1是中文檔，使用encoding解決編碼問題
    # 讀取檔案一律用此格式

print(texts)
# 印出 50 個 "="


```

`readlines`

```python
# 一行一行讀取
with open(file = "../Module2/2-1.py", mode = "r", encoding="utf-8") as f:
    text_list = f.readlines()
print(text_list)
```

```python
# 資料處理
while '\n' in text_list:
    text_list.remove('\n')

result = []
for text in text_list:
    result.append(text.strip("\n"))

print(result)
```

##### 寫入檔案 write( )

```python
# 文字內容
text = '''
傑出的人會努力規劃自己的人生，普通的人則是隨便命運怎麼安排   
一個是充實、有目的的過每一天，一個是隨興的過一天算一天
創業這條路 Michael E.Gerber 
'''

# 寫入檔案
with open(file = "demo.txt", mode = "w", encoding="utf-8") as f:
    f.write(text)
```

```python
# 寫入資料到最後一行 (附加文字在檔案內容最後一行)
# People say that success just happens to you. It doesn't!
t = "People say that success just happens to you. It doesn't!"
with open (file = "demo.txt", mode = "a", encoding="utf-8") as f:
    f.write(t+'\n') # 換行符號
```
