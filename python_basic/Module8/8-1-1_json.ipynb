### `json` 套件

| 用法 | 說明 |
|:--:|:--:|
|json.dumps(<list, dict>)|物件轉JSON並寫入文件|
|json.dump(<list, dict>)|物件轉JSON字串|
|json.loads()|讀取JSON文件|
|json.load()|將JSON字串轉成Python物件|

```python
#引入 json
import json

fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange'] 
    },
    "kiwi": {
        'price': 150,
        'quantity': 30,
        'color': ['orange'] 
    }
}
# json 格式 -> 只能是list, dict
with open("fruit.json", mode="w")as f:
    f.write(json.dumps(fruit_dict)) 
    # f.write 只能寫入str，會造成接收的人讀取錯誤，所以使用json.dump檢查資料有沒有符合list, dict格式。
```

`json.dump(<list, dict>, <file>)`
是常用寫法

```python
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange'] 
    },
    "kiwi": {
        'price': 150,
        'quantity': 30,
        'color': ['orange'] 
    }
}

with open("fruit2.json", mode="w")as f:
    json.dump(fruit_dict, f, indent=4) 
    # 將物件轉json字串
    # indent 空格，用來整理格式
```

```python
# loads()
with open("fruit2.json", "r", encoding="utf-8") as f:
    data = f.read()

print(type(data))
fruit_dict = json.loads(data) #讀取字串物件
print(type(fruit_dict))
print(fruit_dict)
```

```python
# load()
with open("fruit2.json", "r", encoding="utf-8") as f:
    data = json.load(f)

print(type(data))
print(data)
```
