整數 (int)、浮點數 (float)

```python
num = 10
floatnum = 10.0
print(num, floatnum)
print(type(floatnum))
```

布林值 (boolean)

```python
x = True
y = False
print(x, y)
print(type(x))
```

字串 (string)

```python
date = "2026/3/27"
print(date)
print(type(date))
```

列表 (list)

```python
listnums = [1, 2, 5, 9]
print(listnums)
print(type(listnums))
```

元組 (tuple)

```python
tupnums = (1, 2, 3, 4)
print(tupnums)
print(type(tupnums))
```

集合 (set)

```python
setnums = {1, 2, 3, 4}
print(setnums)
print(type(setnums))
```

字典 (dict)

```python
dictnum = {"one":1, "two":2, "three":3}
print(dictnum)
print(type(dictnum))
```

```python
# 創建一個空的 dict
emptydict = {}
print(emptydict)
print(type(emptydict))
```

```python
# 創建一個空的list
emptylist = []
print(emptylist)
print(type(emptylist))
```

```python
# 創建一個空的 set
data = set()
print(data)
print(type(data))
```
