算術運算子

```python
print(3+5)
# 相除取餘數
print(10%3)
# 相除取整數
print(10//3)
# 平方
print(10**3)
```

比較運算子 (用 if 判斷)

```python
print(5 == 5)
print(5 != 5)
print(5 < 5)
print(5 > 5)
print(5 <= 5)
print(5 >= 5)

score = 55
passing_score = 60
print(f"you got {score} scores.")
if score >= passing_score:
    print("you passed!")
elif score < passing_score:
    print("you're not passed!")
```

邏輯運算子 and / or / not

```python
username = "david"
password = "1234"

if username == "david" and password == "12345":
    print("登入成功")
if username == "david" or password == "12345":
    print("帳號或密碼之一正確")
if username != "david" and password != "12345":
    print("帳號密碼錯誤")

```

成員運算子 in / not in

```python
mylist = [9, 5, 2, 7]
if 2 in mylist:
    print("yes")
if 8 not in mylist:
    print("no")
```

身分運算子 is / is not

```python
x = 123
y = 123
print(id(x)) # 檢查記憶體位置
print(id(y))
print(x is y)

x = [1, 2, 3]
y = [1, 2, 3]
print(id(x)) # 檢查記憶體位置
print(id(y))
print(x is y) #記憶體位置不同
```
