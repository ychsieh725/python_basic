| 用法 | 說明 |
|:---:|:---:|
| `range(m)` | 走訪 0 到 m-1 的數字 |
| `range(n, m)` | 走訪 n 到 m-1 的數字  |
| `range(n, m, step)` | 以step為間距，走訪 n 到 m-1 的數字 |

##### for

```python
# for <變數> in <資料>
for char in "contents":
    # print(char)
    print(char, end = ", ") #預設是給\n換行符號
```

資料來源：list

```python
data = [1, 2, 3, 4, 5]
for num in data:
    if num >3:
        print(num, end = ",")
```

```python
# range(m) -> 走訪 0 到 m-1 的數字
for i in range(5):
    print("Hello World", end = ",")
```

```python
# range(m) -> 走訪 0 到 m-1 的數字
for i in range(5):
    print(i, end = ",")
```

```python
# range(n, m) -> 走訪 n 到 m-1 的數字
for i in range(1, 11):
    print(i, end = ",")
```

```python
# range(n, m, step) -> 以step為間距，走訪 n 到 m-1 的數字
for i in range(2, 11, 2):
    print(i, end = ",")
```

### 小試身手 for  
請產出 a 開頭的 n 個偶數(含 a 本身)  
Ex:  
a = 5, n = 3  
Output : 6 8 10

a = 4, n = 5  
Output : 4 6 8 10 12

我的解法

```python
a = 5
n = 3

if a % 2 != 0:
    a += 1

for i in range(a, a+(n*2), 2):
    print(i, end = ",")
```

老師解法

```python
a = 5
n = 3

for i in range(n):
    if a % 2 == 0:
        print(a+i*2)
    else:
        print(a+1+i*2, end = ",")
```

```python
x = "* "
y = " "
for i in range(5):
    print(f"{y*i}{x*(5-i)}")

for i in range(1, 6):
    print(f"{y*(5-i)}{x*i}")
```
