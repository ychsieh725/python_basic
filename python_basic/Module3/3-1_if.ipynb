條件語句 if-elif-else

```python
'''
條件判斷
if ...
if ... else ...
if ... elif ... else ...

'''
```

##### if

```python
num = 100

if num > 50:
    print("大於")
```

##### if...else...

```python
num = 30

if num > 50:
    print("大於")
else:
    print("小於")
```

##### if...elif... (else)

```python
fruit = "banana"

if fruit == "apple":
    print("it's apple.")
elif fruit == "grape":
    print("it's grape.")
elif fruit == "banana":
    print("it's banana.")
```
