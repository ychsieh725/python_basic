##### break

```python
'''
當偵測到字母 t 時，就會強制結束迴圈
'''
texts = "contents"
for char in texts:
    print(char)
    if char == "t":
        break
```

##### continue

```python
'''
當偵測到字母 t 時，
會跳過本次迴圈剩下的程式碼 print(char)，
但不會結束迴圈，仍然會進入下一圈繼續執行
'''
texts = "contents"
for char in texts:
    if char == "t":
        continue #跳過 continue 以下的程式碼 進到下一個迴圈
    print(char) #放在 continue 以下
```

計算 1 到 100 之間的奇數之和，但要跳過所有被 5 整除的數
