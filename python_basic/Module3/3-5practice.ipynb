題目：自動折扣結帳機
假設你有一份購物清單，裡面存放著每件商品的價格：
prices = [120, 50, 200, 80, 150]
請你寫一段程式碼，達成以下目標：
使用迴圈計算這份清單的總金額。
進階條件：如果商品的價格超過 100 元，該商品就打 9 折（乘以 0.9）；如果低於 100 元則維持原價。
最後印出調整後的總金額。
進階思考題
如果老闆今天說：「為了慶祝開幕，只要購物籃裡出現任何一個超過 150 元的商品，全單直接再減 20 元（只需減一次），但如果沒出現就不減。」

```python
prices = [120, 50, 200, 80, 150]
total = 0
for i in prices:
    if i > 100:
        i = i * 0.9
    total += i
for i in prices:
    if i > 150:
        total -= 20  
        break    
print(int(total))
```

```python
prices = [120, 50, 200, 80, 150]
has_expensive_item = False
total = 0

for i in prices:
    
    if i > 100:
        i = i * 0.9
        if i > 150:
            has_expensive_item = True
    total += i

if has_expensive_item == True:
    total -= 20
print(int(total))
```
