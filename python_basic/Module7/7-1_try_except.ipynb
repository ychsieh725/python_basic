##### try...except...

```python
# 可以連續定義數個例外處理機制
# 最符合的例外事件類別會先拋出
try:
    x = 100
    print(y) # 發生錯誤就跳到 except 繼續執行
    print(x)
except:
    print("錯誤")
```

```python
try:
    x = 100
    print(y) # 發生錯誤就跳到 except 繼續執行
    print(x)
# as -> 賦值
except Exception as e: # except 後可接錯誤類別
    print(f"發生錯誤:{e}")
```

```python
# 多層例外處理:
try:
    x = 100
    print(y) 
    print(x)
except NameError as e: # 指定錯誤類別
    print(f"發生NameError:{e}")

except Exception as e: # 涵蓋多數錯誤類別，放在最下面
    print(f"發生錯誤:{e}")
```

##### try...except...else...finally

```python
# 如果沒有錯誤，就會執行 else 程式區塊
# finally 程式區塊是無論有無錯誤，都會執行

try:
    x = 10
    print(z)

except Exception as e:
    print(e)

else:
    print("沒有錯誤發生，進到else")

finally:
    print(f"無論如何都要執行：x = {x}")
```
