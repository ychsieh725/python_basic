`基於既有資料或資訊來使用底下的method，才建立class`

```python
# 類別名稱
class Car:
    # 類別屬性 (Class Attribute)
    wheels = 4

    # 建構函式
    def __init__(self, brand, model):
        self.brand = brand #實例屬性
        self.model = model

    # 方法 (Method)
    def show_info(self):
        print(f"{self.brand} has model - {self.model} and has {Car.wheels} wheels.")


#建立實例
car1 = Car("Toyota", "Altis")
print(car1.model)
car1.show_info()

car2 = Car("Benz", "C300")
car2.show_info()
```
