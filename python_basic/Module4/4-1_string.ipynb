```python
# 字串變數初始化
name = "Alex" #賦值
print(name)

name = "David" #重新賦值，非更改字元
print(name)
```

```python
# 替換字串
# 將 string 中的 str1 替換成 str2
name = "Alex" 
newname = name.replace("Alex", "David")
print(newname)
```

```python
'''
string.strip()
去除字串 string 左、右兩邊的空格
'''
# 去除兩側空格
data = "   Pochita    "
data = data.strip()
print(data)

#去除特定字元
data = "xxxxxPochitaxxxxx"
data = data.strip("xxxxx")
print(data)
#去除左邊

data = "xxxxxPochitaxxxxx"
data = data.lstrip("xxxxx")
print(data)

#去除右邊
data = "xxxxxPochitaxxxxx"
data = data.rstrip("xxxxx")
print(data)
```

```python
# 字串小寫變大寫
'''
str.upper()
將字串 string 裡的字母全部改成大寫
'''
car = "toyota"
car = car.upper()
print(car)
'''
str.lower()
將字串 string 裡的字母全部改成小寫
'''
car = "MITSUBISHI"
car = car.lower()
print(car)
```

```python
# 字串分割成 list
greeting = "hello world! hello John"

data = greeting.split() #空白分割
print(data)

data = greeting.split("!") #指定特定字母分割
print(data)

#字串合併
greeting = " ".join(data)
print(greeting)

#tuple合併
greeting = "-".join(("1234", "56789"))
print(greeting)
```
