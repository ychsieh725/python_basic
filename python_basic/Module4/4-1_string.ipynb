```python
# 字串變數初始化
name = "Alex" #賦值
print(name)

name = "David" #重新賦值，非更改字元
print(name)
```

x = string.replace()

```python
# 替換字串
# 將 string 中的 str1 替換成 str2
name = "Alex" 
newname = name.replace("A", "B")
print(newname)
```

```python
'''
string.strip()
去除字串 string 左、右兩邊的字元
'''
# 去除兩側空格
data = "   Pochita    "
data = data.strip()
print(data)

#去除特定字元
data = "xxxxxPochitaxxxxx"
data = data.strip("xxxxx")
print(data)
#去除左邊

data = "xxxxxPochitaxxxxx"
data = data.lstrip("xxxxx")
print(data)

#去除右邊
data = "xxxxxPochitaxxxxx"
data = data.rstrip("xxxxx")
print(data)
```

```python
# 字串小寫變大寫
'''
str.upper()
將字串 string 裡的字母全部改成大寫
'''
car = "toyota"
car = car.upper()
print(car)
'''
str.lower()
將字串 string 裡的字母全部改成小寫
'''
car = "MITSUBISHI"
car = car.lower()
print(car)

'''
str.title()
將字串 string 裡每個字的字首改成大寫，其餘小寫
'''
greeting = 'hello world, hello john.'
big_greeting = greeting.title()
print(big_greeting)
```

`字串分割成list str.split()`

```python
greeting = "hello world! hello John"

data = greeting.split() #空白分割
print(data)

data = greeting.split("!") #指定特定字母分割
print(data)
```

`list 合併成string：string "<中介>".join(<list>)`

```python
#字串合併
greeting = ['hello', 'world!', 'hello', 'John']
msg = " ".join(greeting)
print(msg)
```

`tuple 合併成string："<中介>".join(<tuple>)`

```python
# tuple合併
nums = ("0921", "123", "456")
phonenum = "-".join(nums)
print(phonenum)
```
