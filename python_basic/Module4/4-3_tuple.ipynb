```python
# tuple 初始化: 第一種
data = ("AIPE", "BDSE")
print(type(data))
print(data)
```

```python
# tuple 初始化: 第二種
data = "AIPE", "BDSE" #沒有括號
print(type(data))
print(data)
```

```python
# 透過指定索引輸出值
data = ("AIPE", "BDSE")
print(data[0]) #index
```

```python
# 複數變數修改值
data = ("AIPE", "BDSE")
print(data)

a, b = data #tuple特性，把裡面的東西拿出來
print(a)
print(b)

# 同理 list 的 enumerate()
```

```python
# tuple 不可以修改指定索引的值
data = ("AIPE", "BDSE")
data[0] = "AIOT"
print(data)
```

```python
# list(tuple)可以修改tuple指定索引的值
data = ("AIPE", "BDSE")
print(data)
data = list(data) #強制轉型
data[0] = "AIOT"
print(data)
```

```python
# 用 for 迴圈逐一輸出資料
data = ("AIPE", "BDSE")
for i in data:    
    print(i)
```

```python
# 用 len 計算 tuple 資料個數
data = ("AIPE", "BDSE")
print(len(data))
```
