```python
# 初始化 dict
fruit_price = {"apple": 50, "orange": 30, "tomato": 70}
print(fruit_price)
```

```python
fruit_price = {
    "apple": 50,
    "orange": 30, 
    "tomato": 70
}
print(fruit_price)
```

```python
# 印出 蘋果的價格 -> 取得某 key 的 value
print(fruit_price["apple"])
#直接用key當index
```

```python
# 修改 橘子的價格
fruit_price["orange"] = 200
print(fruit_price["orange"])
```

```python
# 刪除 番茄
fruit_price = {
    "apple": 50,
    "orange": 30, 
    "tomato": 70
}
fruit_price.pop("tomato")
print(fruit_price)
```

```python
# 建立一個 value 含有 dict 的 dict
info_data = {
    "name": "David",
    "age": 25,
    "info": {
        "nickname": "YC",
        "address": "Taipei",
        "phonenums": [
            "0999999999",
            "0988888888",
            "0977777777"
        ]
    }
}
print(info_data)
```

```python
# 取得 nickname
print(info_data["info"]["nickname"])
```

```python
# 取得所有 phone_number
for nums in info_data["info"]["phonenums"]:
    print(nums)
```

```python
'''
改寫 fruit_dict 每個水果需要 price, quantity/amount, color
apple 有 綠色 和 紅色
orange 有 橘色
tomato 有 紅色
'''

fruit_dict = {
    "apple": {
        "price": "10$", 
        "quantity": 20, 
        "color": ["green", "red"]
    },
    "orange": {
        "price": "10$", 
        "quantity": 20, 
        "color": ["orange"] #就算只有一筆資料也用list保持結構一致
    },
    "tomato": {
        "price": "10$", 
        "quantity": 20, 
        "color": ["red"]
    }
}
print(fruit_dict)

fruit_dict2 = {
    "kiwi": {
        "price": "10$", 
        "quantity": 20, 
        "color": ["green", "yellow"]
    }
}
print(fruit_dict2)
```

##### dict 的合併

```python
# dict 的合併 dict.update()
fruit_dict.update(fruit_dict2)
print(fruit_dict)
```

##### 另外的 print 方式 `pprint` (pretty-print) 

```python
# 從 Module 引入 function
from pprint import pprint
pprint(fruit_dict)
print("="*20)
# 預設不會排序
pprint(fruit_dict, sort_dicts=False)
# sort_dicts 預設True, 不想改就False
```
