```python
# 初始化一個 list
nums = [1, 2, 3]
print(nums)
```

```python
# 新增元素在 list 尾端
# list.append(<data>)
nums = [1, 2, 3]
nums.append(4) #直接改變，不需要變數
print(nums)
```

```python
# 修改索引位置的元素
# list[index] = <值>
nums = [10, 20, 30]
print(nums[2], nums[0]) #以[]使用index

nums[2] = 100 #修改index[2]的值
print(nums)
```

```python
# 刪除索引位置的元素
# list.pop(<index>)
nums = [10, 20, 30]
nums.pop(1)
print(nums)
```

```python
# 在指定索引1新增元素 9 ，其餘元素往後移
# list.insert(<index>, <value>)
nums = [10, 20, 30]
nums.insert(1, 9)
print(nums)
```

```python
# 移除指定的值
# list.remove(<value>)
nums = [10, 20, 30]
nums.remove(10)
print(nums)
```

```python
nums = [10, 20, 30]
for i in range(5):
    nums.append(9)
print(nums)

nums.remove(9) #只會移除一個9
print(nums)

while 9 in nums: #當9在時重複執行
        nums.remove(9)

print(nums)
```

```python
# 計算 list 內的[某個值] 共有幾個
# list.count(<value>)
nums = [10, 20, 30]
for i in range(5):
    nums.append(9)
print(nums)

print(f"資料有{nums.count(9)}個9。")
```

```python
# 計算 list 裡總共有幾個元素
# len()
nums = [10, 20, 30]
print(len(nums))
```

### List comprehension 列表推導式
`new_list = [expression for item in iterable [if condition]]`  
  
`list的名稱 = [表達式 for 元素 in 迭代對象 [if 判斷式(...的條件)]]`

```python
# 原本寫法
nums = []
for i in range(1,21):
    nums.append(i)

print(nums)
```

```python
# 簡化寫法(基於C語言)
nums = [i for i in range(1,21)]
'''
第一個i是 nums.append(i)
第二個i是 for i in range(1, 21):
'''
print(nums)
```

```python
# 只列出5的倍數寫法
nums = []
for i in range(1,21):
    if i % 5 == 0:
        nums.append(i)

print(nums)
```

```python
# 只列出5的倍數簡化寫法
nums = [i for i in range(1,21)if i % 5 == 0]
print(nums)
```

```python
#條件加分寫法
grade = 40
if grade < 60:
    grade += 10
else:
    grade += 5
print(grade)
```

```python
#條件加分三元運算寫法
grade = 40
grade = grade + 10 if grade < 60 else grade + 5
'''
拆解：
1. grade + 10 if grade < 60
2. else grade + 5
'''
print(grade)
```

```python
#及不及格三元運算寫法
grade = 40
passing = "不及格" if grade < 60 else "及格"
print(passing)
```

```python
grades = [70, 88, 45, 94, 55, 20, 75, 63]
passing = 0
notpassing = 0
for grade in grades:
    if grade >= 60:
        passing += 1
    else:
        notpassing += 1
print(passing)
print(notpassing)
```

```python
# 上面簡化寫法
grades = [70, 88, 45, 94, 55, 20, 75, 63]
passing = ["及格" if grade >= 60 else "不及格" for grade in grades]

print(passing)
print(passing.count("及格"))
print(passing.count("不及格"))
```
