```python
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange'] 
    }
}
print(fruit_dict)
```

```python
# 使用 for 迴圈讀取 dict 資料1 
for key in fruit_dict: #只取key
    print(key)
```

```python
# 使用 for 迴圈讀取 dict 資料2
for value in fruit_dict.values(): 
    print(value)
```

必須要知道的方法，key value 都可以取

```python
# 使用 for 迴圈讀取 dict 資料3 -> dict.items()
for i in fruit_dict.items():
    print(i) # 產生tuple，可以取用 
    
for key, value in fruit_dict.items(): #用tuple的方法取得 key, value 的值
    print(key)
    print(value)
```

```python
# 取得 dict 的 value
msg = ""
for key, value in fruit_dict.items():
    # print(value["color"])
    for color in value["color"]:
        msg += color + " "
print(msg)
```

```python
# 練習:所有數量的加總
total = 0
for key, value in fruit_dict.items():
    print(value["quantity"])
    total += value["quantity"]

print(total)
```

```python
#最貴的水果是誰 多少
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange'] 
    },
    "kiwi": {
        'price': 150,
        'quantity': 30,
        'color': ['orange'] 
    }
}
top_price = 0
top_fruit = ""

for fruit, info in fruit_dict.items():
    # print(fruit)
    # print(info["price"])
    if info["price"] > top_price:
        top_price = info["price"]
        top_fruit = fruit

print(top_price)
print(top_fruit)
```

```python
# 資料處理
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]
```

```python
# 將資料整理成 {'apple': ['red', 'green'], 'grape': ['purple', 'green'], 'kiwi': ['yellow', 'green']}
result = {}

for data in fruit_color_list:
    # print(data)

    for fruit, color in data.items():
        if result.get(fruit) == None:
            result[fruit] = [color] #建立這個 key 裡面放 list[color]
        else:
            result[fruit].append(color) #易錯 #已經有的 key 用 append 放進新的顏色
print(result)
```

##### `defaultdict(<type>)`

```python
# from 模組 import 物件
from collections import defaultdict


# 當訪問的 key 不存在時，可以自動創建 key 和預設值<0>，而不是拋出KeyError。
data = defaultdict(int) #預設值為int

# 以int型態的預設值 0 來初始化
print(data["apple"])
print(data)
```

```python
result = defaultdict(list) #預設值為[]
print(result["apple"])
print(result)
```

```python
# 將資料整理成 {'apple': ['red', 'green'], 'grape': ['purple', 'green'], 'kiwi': ['yellow', 'green']}
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]
result = defaultdict(list) #預設值為[]
# print(result["apple"])
# print(result)
for fruit_color in fruit_color_list:
    for fruit, color in fruit_color.items():
        if color not in result[fruit]: #顏色不在 list 中才 append
            result[fruit].append(color) # 呼叫到該key的同時，value 創建一個 [] ，所以可以直接 append color

print(result)
print(result["apple"])
```
