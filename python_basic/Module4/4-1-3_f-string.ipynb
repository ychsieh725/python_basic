```python
# f-string (又稱作 formatted string literals, version >= 3.6)
name = "David"
print(f"Welcome {name}.")

age = 25
print(f"{name} is {age} years old.")
```

```python
# 靠左對齊(右側空格)、靠右對齊(左側空格)
name = "David"
result = f"Name: [{name:<10}], Name: [{name:>10}]"
print(result)
```

```python
# 靠左對齊、靠右對齊，字數不足，補 <字元>
# (沒寫補什麼，預設空格)
name = "David"
result = f"Name: [{name:X<10}], Name: [{name:X>10}]"
print(result)
```

```python
# 置中對齊（左右均分）
name = "David"
result = f"Name: [{name:^10}]"
print(result)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
name = "David"
result = f"Name: [{name:.2}]"
print(result)
```

去識別化應用

```python
user_id = "A112564877"

result1 = f"{user_id:X<10}" #用 X 補滿字元不足10個的 f 字串
print(result1)

result2 = f"{result1:X<10.4}"
print(result2)
```
