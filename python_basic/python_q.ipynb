### 有一筆學生資料如下:

```python
students_grade = [
    {"name": "Denji", "國文": 30, "英文": 10, "數學": 20},
    {"name": "Power", "國文": 60, "英文": 85, "數學": 70},
    {"name": "Aki", "國文": 95, "英文": 88, "數學": 92},
    {"name": "Pochita", "國文": 50, "英文": 65, "數學": 20},
    {"name": "Makima", "國文": 100, "英文": 98, "數學": 95},
    {"name": "Reze", "國文": 65, "英文": 40, "數學": 30},
]
```

##### 1. 印出所有學生姓名

```python
for student_name in students_grade:
    print(student_name["name"])
```

##### 2. 印出 `Pochita` 的數學成績

```python
for i in students_grade:
    if i["name"] == "Pochita":
        print(i["數學"])
```

#### 3. 計算每位學生的總分
輸出格式  
`Phochita 總分: xxx 分`

```python
for i in students_grade:
    sum = 0
    name = i["name"]
    sum += i["國文"] + i["英文"] + i["數學"]
    print(f"{name}總分：{sum}分")
```

#### 4. 計算每位學生的平均分數
輸出格式  
`Pochita 平均分數: xxx 分`

```python
for i in students_grade:
    average = 0
    name = i["name"]
    average += int((i["國文"] + i["英文"] + i["數學"])/3)
    print(f"{name}平均分數：{average}分")
```

#### 5. 找出數學最高分的學生

```python
math_score = 0
top_student = ""
for i in students_grade:
    if i["數學"] > math_score:
        top_student = i["name"]
        math_score = i["數學"]
print(f"最高分的是 {math_score} 分的 {top_student} 同學。")
```

#### 6. 計算全班國文平均

```python
chinese_score = 0
student_num = len(students_grade)
for i in students_grade:
    chinese_score += i["國文"]
print(chinese_score/student_num)
```

#### 7. 找出不及格的學生（任一科 < 60）

```python
for i in students_grade:
    if i["國文"] < 60 or i["數學"] < 60 or i["英文"] < 60:
        print(i["name"])
```

#### 8. 新增「總分」到每個 dict
例如 :  
`{"name": "Pochita", "國文": 80, ..., "總分": 245}`

```python
for i in students_grade:
    sum = 0
    sum += i["國文"] + i["英文"] + i["數學"]
    i["總分"] = sum
    print(i)
```

#### 9. 找出三科都 > 80 的學生

```python
for i in students_grade:
    if i["國文"] > 80 and i["數學"] > 80 and i["英文"] > 80:
        print(i["name"])
```

#### 10. 將資料轉成以下格式 (dict of dict)

```python
"""
{
    "Denji": {"國文": 30, "英文": 10, "數學": 20},
    .....
}
"""
from pprint import pprint
dictofdict = {}
for i in students_grade:
    indict = {}
    sum = 0
    sum += i["國文"] + i["英文"] + i["數學"]
    i["總分"] = sum
    indict["國文"] = i["國文"]
    indict["英文"] = i["英文"]
    indict["數學"] = i["數學"]
    indict["總分"] = sum
    dictofdict[i["name"]] = indict
pprint(dictofdict, sort_dicts=False)
```
