```python
# 建立 function 1 -> 沒有 return值 也沒有參數
"""
某個物件才能使用的 function -> method
e.g. list.append() 可被稱為 list 的 method
"""
def greeting():
    print("hello, world!")

"""
自己建立的只能叫做 function
所以 function != method
"""

greeting()
```

```python
# 建立 function 2 -> 有 return值 沒有參數
def get_greeting_str():
    msg = "hello" + " " + "world!" # 處理字串
    return msg # 回傳處理好的字串

result = get_greeting_str() 
# 有 return 值要用變數承接
print(result)
```

```python
# 建立 function 3 -> 有設定參數
def greeting(name):
    print(f"hello, {name}!")

greeting("David")
greeting("Jerry")
```

```python
# 建立 function 4 -> 預設參數值
def greeting(name, greeting = "hello"): # 要預設參數值的放後方
    print(f"{greeting}, {name}")

greeting("David")
greeting("Jerry", "good morning")
greeting(greeting = "good morning", name = "Jerry") # 指定參數
```

```python
# 改寫 Module2 practice
def get_discount(price):
    if price >= 2000:
        price -= 200*(price//2000)
        print(f"折扣後金額是{price}元。")

get_discount(12800)

print("="*20)

customer_cost = [8000, 9400, 7480]
for cost in customer_cost:
    get_discount(cost)
```

```python
def addargs(*args, **kwargs): #重點是[*]
    print(kwargs)
    total = 0
    for i in args:
        total += i
    return total
total = addargs(5, 10, 15, 20, 25, 30, unit = "元")
print(total)
```
