### lambda 匿名函式  
用法  

`function_name = lambda arguments: expression`

````python
# 建立函式
def add(x, y):
    return x+y

result = add(3, 7)
print(result)
````

````python
# 基本用法
add = lambda x, y: x + y
result = add(3, 7)
print(result)
# lambda 只能用在簡單計算
````

````python
# 條件判斷
def abs(x):
    if x > 0:
        return x
    else:
        return -x
result = abs(-8)
print(result)
````

````python
# 三元運算 -> <if True 回傳的結果> if <判斷式> else <else 要回傳的結果>
abs = lambda x: x if x >= 0 else -x
result = abs(-8)
print(result)
````

### 小試身手  
請建立一個 list ，並在該 list 內產生多個 `[1, 2, x]` 的 list (x 的值要遞增)  
x 到 10 即可  
Ex :  
```Python
[ [1, 2, 1], [1, 2, 2], [1, 2, 3]...[1, 2, 10] ]
```

````python
# 錯誤一
result = []
inside_list = [1, 2]
for i in range(1, 11):
    inside_list.append(i)

print(inside_list)
````

````python
# 錯誤二
# 記憶體位置的錯誤
result = []
inside_list = [1, 2, 0]
for i in range(1, 11):
    inside_list[2] = i
    result.append(inside_list)
print(result)
````

````python
# 錯誤二解法
result = []
inside_list = [1, 2, 0]
for i in range(1, 11):
    inside_list[2] = i
    result.append(inside_list.copy()) # .copy()
print(result)
````

````python
result = []
for i in range(1, 11):
    result.append([1, 2, i])
print(result)
````

````python
# 補充
result = [[1, 2, i] for i in range(1, 11)]
print(result)
````

##### `sort()` 
大小排序

````python
result = []
for i in range(1, 11):
    result.append([1, 2, i])
print(result)

result.sort() #預設由小排到大
print(result)

result.sort(reverse=True) #由大排到小
print(result)
````

每三個 list 中間的值要改成 5  
Ex:  
```python
[ [1, 2, 1], [1, 2, 2], [1, 5, 3], [1, 2, 4], [1, 2, 5], [1, 5, 6]...]
```

````python
result = []
for i in range(1, 11):
    if i % 3 == 0:
        result.append([1, 5, i])
    else:
        result.append([1, 2, i])
print(result)

print("="*100)

result.sort(reverse=True)
print(result)
# 因為第二個數5影響大小順序

print("="*100)

"""
『lambda 真正重要的用法』
key = lambda x: x[2]
"""
result.sort(reverse=True, key=lambda x: x[2]) # lambda return index[2] 來做排序
print(result)
````
