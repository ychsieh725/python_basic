##### 使用 global 讓函式存取全域變數的值

```python
# 全域變數
name = "David"

# 定義函式
def get_name():

    global name
    name = "Sandy" # 區域變數
    
    return name

# 執行函式
data = get_name()

# 輸出 name
print(name)
```
